•	computes an output image where each pixel of coordinates (x,y) contains the absolute difference of the corresponding pixels on M1 and M2:

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the images
image1_path = 'new.jpg'  # replace with your image path
image2_path = 'neu.jpg'  # replace with your image path

# Read the images
M1 = cv2.imread(image1_path)
M2 = cv2.imread(image2_path)

# Check if the images were loaded successfully
if M1 is None or M2 is None:
    print("Error: One or both images not found.")
else:
    # Resize both images to the same dimensions (use dimensions of the first image here)
    M2_resized = cv2.resize(M2, (M1.shape[1], M1.shape[0]))

    # Compute the absolute difference between the images
    Out = cv2.absdiff(M1, M2_resized)

    # Convert color from BGR to RGB for matplotlib
    Out_rgb = cv2.cvtColor(Out, cv2.COLOR_BGR2RGB)

    # Display the output image with matplotlib
    plt.imshow(Out_rgb)
    plt.axis('off')  # Hide axes for better visualization
    plt.show()
```

Implement the following feature extraction techniques using HOG:
1)	Preprocess the Data (64 x 128)
2)	Calculating Gradients (direction x and y)
3)	Calculate the Magnitude and Orientation
4)	Calculate Histogram of Gradients in 8×8 cells
5)	Normalize gradients in 16×16 cell
6)	Features for the complete image


```python
!pip install scikit-image 
```

```python
import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog

# Load the grayscale image
image = cv2.imread('new.jpg', cv2.IMREAD_GRAYSCALE)

# Step 1: Resize the image to (64, 128)
image_resized = cv2.resize(image, (64, 128))

# Step 2: Compute gradients using Sobel operator
grad_x = cv2.Sobel(image_resized, cv2.CV_64F, 1, 0, ksize=3)  # Gradient in x direction
grad_y = cv2.Sobel(image_resized, cv2.CV_64F, 0, 1, ksize=3)  # Gradient in y direction

# Step 3: Compute magnitude and orientation of the gradient
magnitude = cv2.magnitude(grad_x, grad_y)  # Gradient magnitude
orientation = cv2.phase(grad_x, grad_y, angleInDegrees=True)  # Gradient orientation (in degrees)

# Step 4: Compute Histogram of Oriented Gradients (HOG)
# Set HOG parameters: cells of 8x8 pixels, 9 orientation bins, and block size of 2x2 cells
hog_features, hog_image = hog(
    image_resized,
    pixels_per_cell=(8, 8),
    cells_per_block=(2, 2),
    orientations=9,
    block_norm='L2-Hys',
    visualize=True
)

# Step 5: Visualize the results using plt.subplot
plt.figure(figsize=(12, 12))

# Gradient in X direction
plt.subplot(3, 3, 1)
plt.imshow(grad_x, cmap='gray')
plt.title('Gradient in X Direction')
plt.axis('off')

# Gradient in Y direction
plt.subplot(3, 3, 2)
plt.imshow(grad_y, cmap='gray')
plt.title('Gradient in Y Direction')
plt.axis('off')

# Magnitude of the gradient
plt.subplot(3, 3, 3)
plt.imshow(magnitude, cmap='gray')
plt.title('Gradient Magnitude')
plt.axis('off')

# Orientation of the gradient
plt.subplot(3, 3, 4)
plt.imshow(orientation, cmap='hsv')
plt.title('Gradient Orientation')
plt.axis('off')

# Resized image
plt.subplot(3, 3, 5)
plt.imshow(image_resized, cmap='gray')
plt.title('Resized Image (64x128)')
plt.axis('off')

# HOG Visualization
plt.subplot(3, 3, 6)
plt.imshow(hog_image, cmap='gray')
plt.title('HOG Visualization')
plt.axis('off')

# Plot the HOG features (first 100 features for visualization)
plt.subplot(3, 3, 7)
plt.plot(hog_features[:100])
plt.title('First 100 HOG Features')

# Make other axes empty
plt.subplot(3, 3, 8).axis('off')
plt.subplot(3, 3, 9).axis('off')

plt.tight_layout()
plt.show()
```
